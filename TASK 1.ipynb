## We have missing values in Age, Embarked and Cabin Column

#### We will handle missing values using mode in Age column

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

#### We will handle Embarked values using mode

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

#### We will drop Cabin because too many values are msissing

```python
df.drop('Cabin', axis=1, inplace=True)
```

### Normallize/Standardize Numerical Features

### Remove Outliers

#### Will use IQR method

```python
def remove_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return df[(df[column] >= lower) & (df[column] <= upper)]

df = remove_outliers('Age')
df = remove_outliers('Fare')
```
