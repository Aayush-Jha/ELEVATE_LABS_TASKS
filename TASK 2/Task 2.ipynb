```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/jhaaa/ELEVATE INTERNSHIP TASKS/DATASET/Titanic-Dataset.csv")
df.head()
```

```python
df.describe()
```

```python
print("Mean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nStandard Deviation:\n", df.std(numeric_only=True))
```

```python
df['Age'].hist(bins=20)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Fare'].hist(bins=20)
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age")
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.show()
```

```python
df_encoded = df.copy()

df_encoded['Sex'] = df_encoded['Sex'].map({
    'male': 0,
    'female': 1
})
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df_encoded.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(
    df[['Survived','Age','Fare','Pclass']]
)

plt.show()
```

## Key Findings

1. Most passengers were between 20-40 years old.
2. Fare distribution is highly right-skewed.
3. Several outliers exist in Fare.
4. Passenger class has some relationship with Fare.
5. Higher fare passengers tended to survive more often.
